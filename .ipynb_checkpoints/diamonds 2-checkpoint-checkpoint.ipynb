# Cleaning and handling the Data 

```python
diamonds.describe()
```

### as we can notice that the X,Y,and Z have zero values which is not possible since they are the length,width and depth of the diamond

```python
print((diamonds['x']==0).sum())
print((diamonds['y']==0).sum())
print((diamonds['z']==0).sum())
print((diamonds['depth']==0).sum())
# so we have 8,7 and 20 zero values in x,y,and z
```

### a zero value means a missing value so we have to replace those values with nan and then drop these null values.

```python
diamonds[['x','y','z']]=diamonds[['x','y','z']].replace(0,np.nan)
```

```python
diamonds.dropna(inplace=True)
```

```python
diamonds.isnull().sum()
```

# converting categorical data into numerical data 

```python
diamonds.head()
```

```python
# We are going to use one-hot-encoding to convert our categorical data into numerical form.
diamonds_1hot=pd.get_dummies(diamonds)
diamonds_1hot.head()
```

```python
diamonds_clean_data= pd.DataFrame(diamonds_1hot)
diamonds_clean_data.head()
```

# Data Scaling

```python
# we are going to use sklearn standard scaler to scale our numerical values 
from sklearn.preprocessing import StandardScaler
standar_scaler= StandardScaler()
numericals=pd.DataFrame(standar_scaler.fit_transform(diamonds_clean_data[['carat','depth','table','x','y','z']]),columns=['carat','depth','table','x','y','z'],index=diamonds_clean_data.index)
numericals.head()
```

```python
diamonds_clean_data_standard= diamonds_clean_data.copy(deep=True)
diamonds_clean_data_standard[['carat','depth','table','x','y','z',]] = numericals[['carat','depth','table','y','z','x']]
```

```python
diamonds_clean_data_standard.head()
```

# creating Trainig and Testing data 

```python
x=diamonds_clean_data_standard.drop(['price'],axis=1)
y=diamonds_clean_data_standard.price
```

```python
from sklearn.model_selection import train_test_split
train_x,test_x,train_y,test_y=train_test_split(x,y,random_state=2,test_size=0.2)
```

# Model Training 

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
LR= LinearRegression()
LR.fit(train_x,train_y)
y_predictions=LR.predict(test_x)
print(f"accuracy:{LR.score(test_x,test_y)}")
print("Mean squared error: {}".format(mean_squared_error(test_y,y_predictions)))
```

```python
from sklearn.model_selection import cross_val_score
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
lin_scores = cross_val_score(LR, x, y,
scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```
