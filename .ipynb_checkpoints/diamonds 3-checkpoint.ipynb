```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
%matplotlib inline 

from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import MinMaxScaler 
from sklearn.model_selection import train_test_split 

from sklearn.linear_model import LinearRegression 
```

```python
diamonds=pd.read_csv('diamonds.csv')
```

```python
diamonds
```

```python
diamonds.drop(['Unnamed: 0'],axis=1,inplace=True)
```

```python
diamonds
```

```python
y=diamonds['price']
x=diamonds.drop('price',axis=1)
```

```python
x
```

```python
print(f"cuts: {len(x['cut'].unique())}")
print(f"colors: {len(x['color'].unique())}")
print(f"clarities: {len(x['clarity'].unique())}")
```

```python
encoder=LabelEncoder()
x['cut']=encoder.fit_transform(x['cut'])
cut_categories={index: label for index, label in enumerate(encoder.classes_)}
cut_categories
```

```python
x['color']=encoder.fit_transform(x['color'])
color_categories={index: label for index, label in enumerate(encoder.classes_)}
color_categories
```

```python
x['clarity']=encoder.fit_transform(x['clarity'])
clarity_categories={index: label for index, label in enumerate(encoder.classes_)}
clarity_categories
```

```python
x
```

```python
scaler=MinMaxScaler()
x=scaler.fit_transform(x)
```

```python
x_train,x_test,y_train,y_test= train_test_split(x,y,train_size=0.8)
```

```python
LR=LinearRegression()
LR.fit(x_train,y_train)
```

```python
print(f"{LR.score(x_test,y_test)}")
```

```python
from sklearn.metrics import mean_squared_error
y_predictions=LR.predict(x_test)
print("Mean squared error: {}".format(mean_squared_error(y_test,y_predictions)))
```

```python
from sklearn.model_selection import cross_val_score
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
lin_scores = cross_val_score(LR, x, y,
scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)
display_scores(lin_rmse_scores)
```
